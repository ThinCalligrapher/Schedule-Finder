# While loops!!!!
## they need their iterator declared outisde of the while loop

```c++
size_t t {10};
while (t < 100) {
    std::cout << typeid(t).name() << "\n";
    t = static_cast<int>(t);
    t += 40;
}
```

```c++
int i {0};
do {
    std::cout << "This is iteration " << i << std::endl;
    i++;
} while (i < 10);
```

# The do while statement

```c++
// It is useful for having a while loop that will always execute at least once!!
// here is its syntax
// for this we will print ask the user for a positive numbe, then tells it to themn
int k{};
do {
    std::cout << "Enter a positive number: ";
    std::cin >> k;
} while (k < 0);
std::cout << k << std::endl;
```
